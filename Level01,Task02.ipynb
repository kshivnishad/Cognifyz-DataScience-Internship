```python
#Explore the distribution of categorical variables like "Country Code," "City," and"Cuisines."
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
country_dist = data["Country Code"].value_counts()
plt.figure(figsize=(10,5))
country_dist.plot(kind="bar")
plt.title("Distribution of Country Codes")
plt.xlabel("Country Code")
plt.ylabel("Count")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
city_dist = data["City"].value_counts() # Distribution of City (Top 10)
plt.figure(figsize=(10,5))
city_dist.head(10).plot(kind="bar")
plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("City")
plt.ylabel("Count")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
cuisine_dist = data["Cuisines"].value_counts()
plt.figure(figsize=(10,5))
cuisine_dist.head(10).plot(kind="bar")
plt.title("Top 10 Cuisines")
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
top_cities = data["City"].value_counts().head(10) # top 10 cities with highest restaurants
print("Top 10 Cities with Highest Number of Restaurants:")
print(top_cities)
plt.figure(figsize=(10,5))
top_cities.plot(kind="bar")
plt.title("Top 10 Cities by Number of Restaurants")
plt.xlabel("Cities")
plt.ylabel("number of restaurants")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# Top 10 Cuisines with Highest Restaurants
top_cuisines = data["Cuisines"].value_counts().head(10)

print("\nTop 10 Cuisines with Highest Number of Restaurants:")
print(top_cuisines)
plt.figure(figsize=(10,5))
top_cuisines.plot(kind="bar")
plt.title("Top 10 Cuisines by Number of Restaurants")
plt.xlabel("Cuisine")
plt.ylabel("Number of Restaurants")
plt.tight_layout()
plt.show()
```
