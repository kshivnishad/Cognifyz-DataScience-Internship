```python
#Visualize the locations of restaurants on a map using latitude and longitude information.
```

```python
import pandas as pd
import folium
data = pd.read_csv("Dataset .csv")   # Replace with your file name
print(data[['Latitude', 'Longitude']].head())
data = data.dropna(subset=['Latitude', 'Longitude'])
map_center = [data['Latitude'].mean(), data['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)
for index, row in data.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=3,
        color='blue',
        fill=True,
        fill_color='blue').add_to(restaurant_map)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv("Dataset .csv")   
data["Aggregate rating"] = pd.to_numeric(data["Aggregate rating"], errors='coerce')
```

```python
city_counts = data["City"].value_counts()
print("Top 10 restaurants name of Cities:")
print(city_counts.head(10))
plt.figure(figsize=(10,5))
city_counts.head(10).plot(kind = "bar")
plt.title("Top 10 Restaurant of the City")
plt.xlabel("City")
plt.ylabel("number of restaurant")
plt.legend()
plt.xticks()
```

```python
city_avg_rating = data.groupby("City")["Aggregate rating"].mean().sort_values(ascending=False)
print("Top 10 Cities by Average Rating:")
print(city_avg_rating.head(10))
plt.figure(figsize=(10,5))
city_avg_rating.head(10).plot(kind='bar', color='green')
plt.title("Top 10 Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
