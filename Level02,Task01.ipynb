```python
#Compare the average ratings of restaurants with table booking and those without
```

```python
import pandas as pd
data = pd.read_csv("Dataset .csv")
print(data.columns)
data['Aggregate rating'] = pd.to_numeric(data['Aggregate rating'], errors='coerce')
average_ratings = data.groupby('Has Table booking')['Aggregate rating'].mean()
print(average_ratings)
```

```python
def has_table_booking(value):
    value = str(value)
    return "Yes" in value
```

```python
data["TableBookingAvailable"] = data["Has Table booking"].apply(has_table_booking)
avg_ratings = data.groupby("TableBookingAvailable")["Aggregate rating"].mean()
print("Average Ratings based on Table Booking Availability:")
print(avg_ratings)
```

```python
labels = ["Without Table Booking", "With Table Booking"]
ratings = [avg_ratings[False], avg_ratings[True]]

plt.figure(figsize=(5,5))
plt.bar(labels, ratings, color=["salmon", "skyblue"])
plt.title("Average Ratings: Table Booking vs No Table Booking")
plt.ylabel("Average Rating")
plt.ylim(0, 5)  
```

```python
def has_online_delivery(value):
    value = str(value)
    return "Yes" in value  
data["Has Online delivery"] = data["Has Online delivery"].apply(has_online_delivery)
delivery_by_price = data.groupby("Price range")["Has Online delivery"].mean() * 100

print("Percentage of Online Delivery by Price Range:")
print("Has Online delivery")
delivery_by_price.plot(kind='bar', color='skyblue')
plt.title("Online Delivery Availability Across Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Percentage of Restaurants Offering Online Delivery (%)")
```

```python
# Analyze the availability of online delivery among restaurants with different price ranges.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Dataset .csv")
data.head()
```

```python
delivery_price = data.groupby("Price range")["Has Online delivery"].value_counts(normalize = True).unstack()*100
delivery_price.columns = ["No Online Delivery (%)", "Has Online Delivery (%)"]
print(delivery_price)
```

```python
sns.set_theme(style = "whitegrid")
delivery_price.plot(kind = "bar",stacked = True ,figsize = (10,6),color = ["#ff9999","#66b3ff"])
plt.title("Online Delivery Availability by Restaurant Price Range", fontsize = 14)
plt.xlabel("Price range", fontsize = 11)
plt.ylabel("Percentage of Restaurants (%)", fontsize = 11)
plt.legend(title = "Delivery Status",loc = "upper left")
plt.tight_layout()
plt.show()
```
