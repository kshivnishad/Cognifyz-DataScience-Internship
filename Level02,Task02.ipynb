Task: Price Range Analysis

```python
# Determine the most common price range among all the restaurants.
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data= pd.read_csv("Dataset .csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data["Price range"].unique()
price_counts = data["Price range"].value_counts()
print(price_counts)
```

```python
price_range = data["Price range"].mode()[0]
print("Price Range:",price_range)
```

```python
total_restaurants = len(data)
most_common_count = price_counts[price_range]
percentage = (most_common_count / total_restaurants) * 100
print(f"Price Range:",price_range)
print(f"Percentage of Restaurants in this Price Range: {percentage:.2f}%")
```

```python
import matplotlib.pyplot as plt
data['Price range'].value_counts().plot(kind='bar')
plt.title("Distribution of Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
avg_rating = data.groupby("Price range",as_index = False)["Aggregate rating"].mean()
print(avg_rating)
```

```python
plt.bar(avg_rating["Price range"],avg_rating["Aggregate rating"],color = "#000080")
plt.xlabel("Price range")
plt.ylabel("Average Rating")
plt.title("Average Rating per Price Range")
plt.xticks()
plt.show()
```

```python
# Identify the color that represents the highest average rating among different price ranges.
```

```python
import pandas as pd
```

```python
data.head()
```

```python
data[["Price range","Aggregate rating","Rating color"]].head() # checking the required columns
```

```python
avg_rating = data.groupby("Price range")["Aggregate rating"].mean()
avg_rating
```

```python
# Finding the Price Range with Highest Average Rating
```

```python
highest_price_range = avg_rating.idxmax()
highest_avg_rating = avg_rating.max()
print("Highest avg rating:",highest_price_range)
print("Highest avg rating:",highest_avg_rating)
```

```python
filtered_data = data[data["Price range"]== highest_price_range]
highest_rating = filtered_data["Rating color"].mode()[0]
highest_rating_color = filtered_data["Rating color"].mode()[0]
print("Highest Rated Price Range:",highest_rating)
print("Color Representing Highest Average Rating:",highest_rating_color)
```
